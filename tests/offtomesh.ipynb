### Check correctness ###

```python
import trimesh
from collections import Counter

def check_mesh_correctness(mesh_file):
    mesh = trimesh.load(mesh_file)
    
    # Check for non-manifold edges
    edges = mesh.edges_sorted
    edge_counts = Counter(map(tuple, edges))
    non_manifold_edges = [edge for edge, count in edge_counts.items() if count > 2]

    # Check for non-manifold vertices
    vertex_neighbors = mesh.vertex_neighbors
    non_manifold_vertices = [i for i, neighbors in enumerate(vertex_neighbors) if len(neighbors) > 2]

    results = {
        'is_watertight': mesh.is_watertight,
        'non_manifold_edges': len(non_manifold_edges) > 0,
        'non_manifold_vertices': len(non_manifold_vertices) > 0,
        'inverted_faces': not mesh.is_watertight
    }
    return results

def summarize_results(results):
    summary = {
        'total_meshes': len(results),
        'watertight': 0,
        'non_manifold_edges': 0,
        'non_manifold_vertices': 0,
        'inverted_faces': 0
    }

    for result in results:
        if result['is_watertight']:
            summary['watertight'] += 1
        if result['non_manifold_edges']:
            summary['non_manifold_edges'] += 1
        if result['non_manifold_vertices']:
            summary['non_manifold_vertices'] += 1
        if result['inverted_faces']:
            summary['inverted_faces'] += 1

    return summary

def process_directory(input_dir):
    results = []
    for root, dirs, files in os.walk(input_dir):
        for file_name in files:
            if file_name.endswith('.obj'):
                mesh_file = os.path.join(root, file_name)
                result = check_mesh_correctness(mesh_file)
                results.append(result)
                print(f"Results for {mesh_file}: {result}")

    summary = summarize_results(results)
    print(f"Summary: {summary}")

input_dir = "./convertedModelNet10"
process_directory(input_dir)
```
